```python
import pandas as pd
# Carga del archivo 
df= pd.read_csv(r"C:/Henry/python/Python/chicago_crimes.csv")

# Verificación de las primeras filas 

df.head(20)


```

```python
# Convertir las columnas 'Date' y 'Updated On' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Updated On'] = pd.to_datetime(df['Updated On'], errors='coerce')

# Convertir las columnas numéricas a tipo adecuado
df['Beat'] = df['Beat'].astype(int)
df['District'] = df['District'].astype(int)
df['Ward'] = df['Ward'].astype(int)
df['Year'] = df['Year'].astype(int)

 
```

```python
# Eliminar filas con valores nulos en la columna 'Community Area'
df = df.dropna(subset=['Community Area'])

# Convertir la columna a tipo entero
df['Community Area'] = df['Community Area'].astype(int)
```

```python
df.isnull().sum()
```

```python
df['Location Description'] = df["Location Description"].fillna('Unknown')
df['X Coordinate'] = df["X Coordinate"].fillna('Unknown')
df['Y Coordinate'] = df["Y Coordinate"].fillna('Unknown')
df['Latitude'] = df["Latitude"].fillna('Unknown')
df['Longitude'] = df["Longitude"].fillna('Unknown')
df['Location'] = df["Location"].fillna('Unknown')
```
